# 🔧 Cell 1: Setup & Import Libraries

```python
# Cell 1: Setup
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

#Optional: widen display
pd.set_option('display.max_columns', None)

# Load data (upload CSV first)
try:
    df = pd.read_csv('KMS_Sql_Case_Study.csv', encoding='utf-8')
except UnicodeDecodeError:
    df = pd.read_csv('KMS_Sql_Case_Study.csv', encoding='latin1')


df.head()
```

# 🔍 EDA CELLS

*📊Cell 2: Summary info*

```python
# Cell 2: Dataset Overview
print(df.shape)
print(df.dtypes)
df.describe()
```

*🧼Cell 3: Missing Values*

```python
df.isnull().sum()
```

*📦 Cell 4: Product Category Sales*

```python
# Cell 4: Total sales by product category
df.groupby('Product Category')['Sales'].sum().sort_values(ascending=False)
```

*🌍 Cell 5: Sales by Region*

```python
# Cell 5: Barplot - Sales by Region
plt.figure(figsize=(10,6))
sns.barplot(x='Region', y='Sales', data=df, estimator=sum, ci=None)
plt.xticks(rotation=45)
plt.title("Total Sales by Region")
plt.show()
```

*🔢 Cell 6: Top 10 Customers by Profit*

```python
# Cell 6: Top 10 profitable customers
df.groupby('Customer Name')['Profit'].sum().sort_values(ascending=False).head(10)
```

*🧑🏽‍💼 Cell 7: Customer Segment Distribution*

```python
# Cell 7: Pie chart of customer segments
df['Customer Segment'].value_counts().plot.pie(autopct='%1.1f%%', figsize=(6,6))
plt.title("Customer Segment Distribution")
plt.ylabel("")
plt.show()
```

*📦 Cell 8: Sales by Sub-Category*

```python
# Cell 8: Sales by sub-category (top 10)
top_subcat = df.groupby('Product Sub-Category')['Sales'].sum().sort_values(ascending=False).head(10)
top_subcat.plot(kind='barh', figsize=(8,5), title='Top 10 Product Sub categories by Sales')
plt.xlabel('Sales')
plt.show()
```

*🛒 Cell 9: Shipping Mode vs Cost*

```python
# Cell 9: Avg Shipping Cost per ShipMode
df.groupby('Ship Mode')['Shipping Cost'].mean().sort_values(ascending=False)
```

*🧾 Cell 10: Profit by Order Priority*

```python
# Cell 10: Boxplot - Profit by Order Priority
plt.figure(figsize=(10,6))
sns.boxplot(x='Order Priority', y='Profit', data=df)
plt.title("Profit Distribution by Order Priority")
plt.show()
```

*🕒 Cell 11: Trend Over Time*

```python
# Cell 11: Sales Over Time
df['Order Date'] = pd.to_datetime(df['Order Date'])
df.set_index('Order Date')['Sales'].resample('M').sum().plot(figsize=(12,6), title='Monthly Sales Trend')
plt.ylabel('Sales')
plt.show()
```

*🏙️ Cell 12: Sales in Ontario Only*

```python
# Cell 12: Sales in Ontario by Category
ontario = df[df['Province'] == 'Ontario']
px.bar(ontario, x='Product Category', y='Sales', color='Product Category', title="Ontario Sales by Product Category")
```

*🧮 Cell 13: Correlation Matrix*

```python
# Cell 13: Correlation matrix (Sales, Profit, Discount, Cost)
numeric_cols = ['Sales', 'Profit', 'Discount', 'Shipping Cost', 'Order Quantity']
sns.heatmap(df[numeric_cols].corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

*🧑🏽‍💻 Cell 15: Top Corporate Customers*

```python
# Cell 15: Top 5 corporate customers by number of orders
corporate = df[df['Customer Segment'] == 'Corporate']
corporate['Order ID'].nunique()
top_corp = corporate.groupby('Customer Name')['Order ID'].nunique().sort_values(ascending=False).head(5)
top_corp.plot(kind='bar', title="Top 5 Corporate Customers by Order Count")
plt.ylabel("Number of Orders")
plt.show()
```

# INTERACTIVE DASHBOARD

*📈 Cell 16: Interactive Bar Chart (Plotly)*

```python
# Cell 16: Interactive plot - Sales by Region
fig = px.bar(df, x='Region', y='Sales', color='Region', title='Interactive: Sales by Region',
             labels={'Sales': 'Total Sales'},
             hover_data=['Province', 'Product Category'])
fig.show()
```

📉 Cell 17: Interactive Time Trend Line

```python
# Cell 17: Interactive sales trend
df['Order Date'] = pd.to_datetime(df['Order Date'])
monthly = df.groupby(df['Order Date'].dt.to_period('M'))['Sales'].sum().reset_index()
monthly['Order Date'] = monthly['Order Date'].dt.to_timestamp()

fig = px.line(monthly, x='Order Date', y='Sales', title='Interactive Monthly Sales Trend')
fig.show()
```
